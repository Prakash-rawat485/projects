
Some credit-related columns are very closely related because they are calculated from each other, so they give almost the same information.

Columns like credit limit, balance, and available credit depend on each other, which means they are not independent and repeat similar data.

The utilization ratio is calculated using balance and credit limit, so it also shows similar information in a different form.

Credit Limit & Avg_Open_To_Buy (1.00): These are perfectly linked. If your credit limit is high, your "open to buy" (available credit) is also high

Total Revolving Balance & Utilization Ratio (0.62): This shows a clear link. Customers who carry a "revolving balance" (unpaid debt) naturally use up more of their available credit limit.


Total Transaction Amount & Total Transaction Count (0.81): This is a very strong connection. It tells us that customers who swipe their cards more often are also the ones spending the most money.

Total Relationship Count & Transaction Amount (-0.35): This suggests that customers who have many different products with the bank (like loans and savings) might actually spend less on their credit cards compared to those who only have a card.

Because these columns are so similar, we may remove some of them later to avoid repeating the same information and to make the model work better.

```python
df['activity_level'] = pd.qcut(df['Total_Trans_Ct'], q=3, labels=['Low','Medium','High'])

sns.countplot(x='activity_level', hue='Attrition_Flag', data=df)
plt.show()
```

Activity is our best early warning system. Customers in the 'Low' activity bracket are at an extremely high risk of leaving, while 'High' activity users almost never churn.

```python

churned = df[df['Attrition_Flag'] == 'Attrited Customer']
existing = df[df['Attrition_Flag'] == 'Existing Customer']

cols = [
    'Customer_Age', 'Credit_Limit', 'Total_Trans_Amt',
    'Total_Trans_Ct', 'Months_Inactive_12_mon',
    'Total_Amt_Chng_Q4_Q1', 'Total_Ct_Chng_Q4_Q1'
]


mean_existing = existing[cols].mean()
mean_churned = churned[cols].mean()


median_existing = existing[cols].median()
median_churned = churned[cols].median()


std_existing = existing[cols].std()
std_churned = churned[cols].std()
summary = pd.DataFrame({
    'Mean_Existing': mean_existing,
    'Mean_Churned': mean_churned,
    'Median_Existing': median_existing,
    'Median_Churned': median_churned,
    'Spread_Existing': std_existing,
    'Spread_Churned': std_churned
})

summary
```
