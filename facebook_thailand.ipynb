```python
# this looks messy so making a single column out of the  the all reactions columns:
```

```python
df['engagement'] = df['num_reactions'] + df['num_comments'] + df['num_shares']+df['num_likes']
```

```python
sns.barplot(x='engagement', y='status_type', data=df)
```

```python
# While the bar plot clearly identifies Video as our top-performing category, the underlying data suggests a "Boom or Bust" pattern. With a mean 
 # of 1,284 but a median of 660, we can see that our "average" is being heavily influenced by a few exceptionally viral posts.
```

```python
# now checking which status type gives the highest engagemnet.
```

```python
engagement_by_type=df.groupby('status_type')['engagement'].agg(['mean', 'median', 'std'])
engagement_by_type
```

**in the above stats we can see that mean is greater for all post than median ,it  tells us that some post becomes so viral which pulls the mean ,hence rigghly skewed. so in this case the mean cannot be trusted fully,we should see median now.**

**when we look the mean we see that video has the highest median,which is also greater than the avg of photo,this implies that the mediocre videos perform much better than the great photos.**

 Conclusion :

From the analysis of the Facebook Thailand dataset, we found clear patterns in user engagement.Photo posts receive the highest engagement compared to links and status updates but somewhere videos are also performing extraordinary but are not consistent.

Engagement also changes based on posting time and day, which means posting at the right time is important. Reactions, comments, and shares are closely related — posts that perform well in one usually perform well in all.

Some posts have extremely high engagement(video), indicating viral content.

Overall, focusing on photo content and posting at the right time can help increase engagement.
